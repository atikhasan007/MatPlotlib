```
import plotly
import pandas as pd
import plotly.express as px
```

```
# Create the pandas DataFrame
df = pd.DataFrame([[20, 10,"A"], [30, 15,"A"], [40, 24,"A"],[50, 10,"A"],[60, 20,"A"],
                   [20, 14,"B"], [30, 18,"B"], [40, 27,"B"],[50, 15,"B"],[60, 25,"B"],
                   [20, 17,"C"], [30, 20,"C"], [40, 30,"C"],[50, 20,"C"],[60, 29,"C"]], columns = ['Weight', 'Height','Section'])
# print dataframe.
df
```

```
fig = px.line(df, x='Weight',y='Height')
fig.show()
```

```
fig = px.line(df, x='Weight',y='Height',color='Section')
fig.show()
```

```
fig = px.line(df, x="Weight",y="Height",width=1000,height=600, color='Section')
fig.show()
```

```
#adding marker
fig = px.line(df, x="Weight",y="Height",width=1000,height=600,markers=True,color='Section')
fig.update_traces(marker_color=12)
fig.show()
```

```
#add data labels
fig = px.line(df, x="Weight", y="Height",
              width=1000, height=600,markers=True, text="Height"
              ,color="Section"

)
fig.update_traces(marker_size=10)
fig.show()
```

```
#customaizing data labels
fig = px.line(df, x="Weight",y="Height",
              width=1000,height=600,markers=True,text="Height",
              color='Section')
fig.update_traces(textposition="top center")
fig.update_traces(marker_size=10)
fig.show()
```

```
#adding chart title
fig = px.line(df, x="Weight",y="Height",
              width=1000,height=600,markers=True,text='Height',color='Section')
fig.update_traces(textposition='top center',marker_size=10)
fig.update_layout(title_text = 'Height Across Weights and Sections.')
fig.show()
```

```
fig = px.line(df,  x='Weight', y='Height',
                 width=1000, height=600,markers=True,text='Height',color='Section')
fig.update_traces(textposition="top center",marker_size=10)
fig.update_layout(title_text='Height Across Weights and Sections', title_x=0.5)
fig.show()
```

```
#Changing Line color
fig = px.line(df,  x='Weight', y='Height',
                 width=1000, height=600,markers=True,text='Height',color='Section',
              color_discrete_map={
        'A': 'aquamarine',
        'B': 'turquoise',
        'C':'aqua'})
fig.update_traces(textposition="top center",marker_size=10)
fig.update_layout(title_text='Height Across Weights and Sections', title_x=0.5)
fig.show()
```

```
#changing theme of the chart
for template_element in ["plotly", "plotly_white", "plotly_dark", "ggplot2", "seaborn", "simple_white", "none"]:
    fig = px.line(df,  x='Weight', y='Height',template=template_element,
                 width=1000, height=600,markers=True,text='Height',color='Section',
              color_discrete_map={
        'A': 'aquamarine',
        'B': 'turquoise',
        'C':'aqua'})
    fig.update_traces(textposition="top center",marker_size=10)
    fig.update_layout(title_text='Height Across Weights and Sections', title_x=0.5)
    fig.show()

```

```
fig = px.line(df,  x='Weight', y='Height',template='plotly_dark',
                 width=1000, height=600,markers=True,text='Height',color='Section',
              color_discrete_map={
        'A': 'aquamarine',
        'B': 'turquoise',
        'C':'aqua'})
fig.update_traces(textposition="top center",marker_size=10)
fig.update_layout(title_text='Height Across Weights and Sections', title_x=0.5)
fig.show()
```
