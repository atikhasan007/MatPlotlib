# stack Plot

```python
x = [1,2,3,4,5]
area1 = [2,3,2,5,3]
area2 = [2,3,4,5,6]
area3 = [1,3,2,4,2]
l  = ["area1","area2","area3"]
plt.stackplot(x,area1,area2,area3,labels = l,colors=["r",'g','y'],baseline="sym")
plt.legend(loc=2)
plt.show()
```

# Fill_Between Plot


```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x,y,color='r')

plt.fill_between(x=[2,4],y1=2,y2=4)


plt.title("python",fontsize=14)
plt.xlabel("x-axis",fontsize=14)
plt.ylabel("y-axis",fontsize=14)
plt.show()
```

```python
import numpy as np
x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])

plt.plot(x,y,color='r')

plt.fill_between(x,y,color='g',where=((x>=2) & (x<=4)))


plt.title("python",fontsize=14)
plt.xlabel("x-axis",fontsize=14)
plt.ylabel("y-axis",fontsize=14)
plt.show()
```

```python
import numpy as np
x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])

plt.plot(x,y,color='r')

plt.fill_between(x,y,color='g',where=((x>=2) & (x<=4)),alpha=0.5)


plt.title("python",fontsize=14)
plt.xlabel("x-axis",fontsize=14)
plt.ylabel("y-axis",fontsize=14)
plt.show()
```
