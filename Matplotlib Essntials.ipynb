```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(11)
y = x ** 2

print(x,y)
```

# Basic Plotting

```python
plt.scatter(x,y,color='red',marker="*")
```

```python
plt.plot(x,y, color='red',marker='*')
#we can add labels and title

plt.xlabel("x axis title")
plt.xlabel("y axis title")
plt.title("Figure/plot Title")
plt.xticks(range(0,11))
plt.yticks(range(0,101,10))
plt.show()
```

# SubPlot and tight_layer

```python
#plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,3,1)
plt.plot(x,y,color='red',marker='*')

plt.subplot(1,3,2)
plt.plot(x,y,color='blue',marker='*')


plt.subplot(1,3,3)
plt.plot(x,y,color='green',marker='*')
plt.show()
```

# Matplotlib Object Oriented approach

```python
fig1 = plt.figure(figsize=(8,5))
# [left, bottom, width, height (range 0 to 1)

axes = fig1.add_axes([0,0,1,1])
axes.plot(x,y,color='red')

axes.set_xlabel("x label")
axes.set_ylabel("y label")
axes.set_title("Figure Title")
axes.set_xticks(range(0,11))
axes.set_yticks(range(0,101,10))
fig1.savefig("graphsave.pdf", dpi=200)
```

```python
fig, axes = plt.subplots(2,2)
```

```python
fig, axes = plt.subplots(2,2)
plt.tight_layout()
```

```python
#we can share x and y axis
fig, axes = plt.subplots(2,2,sharex=True,sharey=True,figsize=(8,5))
plt.tight_layout() #for auto adjustment


axes[0][0].plot(x,y,color='red',marker='*')
axes[0][0].set_xlabel('x label')
axes[0][0].set_ylabel('y lable')
axes[0][0].set_title("Figure Title")

axes[1,1].scatter(x,y);
```

# Let's plot some data

```python
#Tuple unpacking to grab fig and axes
fig, axes = plt.subplots(figsize=(10,5))

#adding data to the plot using axes object
axes.plot(x,y,color='blue',marker='o')


#adding labels and title using axes object
axes.set_xlabel=('x label')
axes.set_ylabel  = ('y label')
axes.set_title('Plot_Title');
```

# Saving figures

```python
#create an empty canvas
fig3 = plt.figure(figsize=(10,5))

#creating main axes - values b/w 0 and 1
axes_main = fig3.add_axes([0,0,1,1])

#creatin inset axes - values b/w 0 and 1
axes_inset = fig3.add_axes([.3,.5,.3,.4])
axes_main.plot(x,y,color='green',marker='o')
axes_main.set_xlabel('x lable_main')
axes_main.set_ylabel('y label_main')

axes_main.set_title('Title_main')


#axes_main.set_facecolor("yellow")
axes_inset.plot(y,x,color='red')
axes_inset.set_xlabel('x label')
axes_inset.set_ylabel('y label')
axes_inset.set_title('Title_inset')

#axes_inset.set_facecolor('green')
fig3.savefig('filename.jpg', dpi=200, bbox_inches='tight')

#DPI = Dots Per Inch  ,     
```

```python
fig, axes = plt.subplots(figsize=(8,6))


axes.plot(x, y,
          color="purple", 
          lw=3,
          ls='--', 
          marker='s', 
          markersize=8,
          markerfacecolor="yellow", 
          markeredgewidth=3,
          markeredgecolor="red");
```

```python
fig, axes = plt.subplots(figsize=(10,8))

axes.plot(x, x+1, color="red", linewidth=0.25)
axes.plot(x, x+2, color="red", linewidth=0.50)
axes.plot(x, x+3, color="red", linewidth=1.00)
axes.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options -, , -., :, steps
axes.plot(x, x+5, color="green", lw=5, linestyle='-')
axes.plot(x, x+6, color="green", lw=5, ls='-.')
axes.plot(x, x+7, color="green", lw=5, ls=':')

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
axes.plot(x, x+9, color="blue", lw=3, ls='-', marker='o', markersize=8, markerfacecolor="yellow")
axes.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
axes.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
axes.plot(x, x+12, color="blue", lw=3, ls='--', marker='1', markersize=20)

# marker size and color
axes.plot(x, x+13, color="purple", lw=3, ls='-', marker='o', markersize=2)
axes.plot(x, x+14, color="purple", lw=3, ls='-', marker='o', markersize=4)
axes.plot(x, x+15, color="purple", lw=3, ls='-', marker='o', markersize=8, markerfacecolor="red")
axes.plot(x, x+16, color="purple", lw=3, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```
