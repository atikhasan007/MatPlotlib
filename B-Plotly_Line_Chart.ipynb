**Importing Libraires**

```
import plotly
import pandas as pd
import plotly.express as  px
```

```
# Create the pandas DataFrame
df = pd.DataFrame([[20, 10], [30, 15], [40, 14],[50, 18],[60, 20]], columns = ['Weight', 'Height'])
# print dataframe.
df
```

```
fig = px.line(df, x="Weight", y="Height", title='Weight Vs Height')
fig.show()
```

```
fig = px.line(df,  x='Weight', y='Height',
                 width=1000, height=600)
fig.show()
```

```
fig = px.line(df,  x='Weight', y='Height',
                 width=1000, height=600,markers=True)
fig.update_traces(marker_size=10)
fig.show()
```

```
fig = px.line(df,  x='Weight', y='Height',
                 width=1000, height=600,markers=True,text='Height')
fig.update_traces(marker_size=10)
fig.show()
```

```
fig = px.line(df,  x='Weight', y='Height',
                 width=1000, height=600,markers=True,text='Height')
fig.update_traces(textposition="top center")
fig.update_traces(marker_size=10)
fig.show()
```

```
fig = px.line(df,  x='Weight', y='Height',
                 width=1000, height=600,markers=True,text='Height')
fig.update_traces(textposition="top center",marker_size=10)
fig.update_layout(title_text='Height Across Weights', title_x=0.5)
fig.show()
```

```
fig = px.line(df,  x='Weight', y='Height',
                 width=1000, height=600,markers=True,text='Height')
fig.update_traces(textposition="top center",line_color='turquoise',marker_size=10)
fig.update_layout(title_text='Height Across Weights', title_x=0.5)
fig.show()
```

```
for template_element in ["plotly", "plotly_white", "plotly_dark", "ggplot2", "seaborn", "simple_white", "none"]:
    fig = px.line(df,  x='Weight', y='Height',
                 width=1000, height=600,markers=True,text='Height',template=template_element)
    fig.update_traces(textposition="top center",line_color='turquoise',marker_size=10)
    fig.update_layout(title_text='Height Across Weights', title_x=0.5)
    fig.show()
```
