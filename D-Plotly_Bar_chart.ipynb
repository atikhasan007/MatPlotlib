```python
#importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import warnings
warnings.filterwarnings('ignore')
```

```python
#declaring Datafreme
# Create the pandas DataFrame
df = pd.DataFrame([['Nick', 10], ['Tom', 15], ['Ram', 14]], columns = ['Name', 'Age'])
# print dataframe.
df
```

```python
fig = px.bar(df, x="Name",y="Age")
fig.show()
```

```python
#adjusting chart size
fig = px.bar(df, x="Name",y="Age",width=500, height=500)
fig.show()
```

```python
#adjusting bar width
fig = px.bar(df, x="Name", y="Age", width=1000,height=600)
fig.update_traces(width=0.25) #Bar width
fig.show()
```

```python
#add Data Labels
fig = px.bar(df, x="Name", y="Age",
             width=1000,height=600, text_auto=True)
fig.update_traces(width=0.5,textfont_size=12, textangle=90,textposition='outside')
fig.show()
```

```python
#adding chart Title
fig = px.bar(df, x='Name', y='Age',
                 width=1000, height=600, text_auto=True)
fig.update_traces(width=0.5,textfont_size=12, textangle=0, textposition="outside")
fig.update_layout(title_text='Age Across Names', title_x=0.5)
fig.show()
```

```python
#changing Bar color
fig = px.bar(df, x='Name', y='Age',
                 width=1000, height=600, text_auto=True)
fig.update_traces(width=0.5,textfont_size=12, textangle=0, textposition="outside",marker_color='red')
fig.update_layout(title_text='Age Across Names', title_x=0.5)
fig.show()
```

```python
#change theme of the chart
for template_element in ["plotly", "plotly_white", "plotly_dark", "ggplot2", "seaborn", "simple_white", "none"]:
    fig = px.bar(df, x='Name', y='Age',
                 width=1000, height=600, text_auto=True, template=template_element)
    fig.update_traces(width=0.5,textfont_size=12, textangle=0, textposition="outside")
    fig.update_layout(title_text='Age Across Names', title_x=0.5)
    fig.update_traces(marker_color='turquoise')
    fig.show()
```

```python
fig = px.bar(df, x='Name', y='Age',
                 width=1000, height=600, text_auto=True, template="plotly_white")
fig.update_traces(width=0.5,textfont_size=12, textangle=0, textposition="outside")
fig.update_layout(title_text='Age Across Names', title_x=0.5)
fig.update_traces(marker_color='turquoise')
fig.show()
```

```python
# Create the pandas DataFrame
df = pd.DataFrame([['Doctor', 30000], ['Consultant', 25000], ['Construction Worker', 15000],
                  ['Software Engineer', 28000],['Chef', 25000] ], columns = ['Occupation', 'Average Salary'])
# print dataframe.
df
```

```python
fig = px.bar(df, x='Occupation', y='Average Salary',
                 width=1000, height=600, text_auto=True, template="plotly_white")
fig.update_traces(width=0.5,textfont_size=12, textangle=0, textposition="outside")
fig.update_layout(title_text='Average Salary Across Occupation', title_x=0.5)
fig.update_traces(marker_color='turquoise')
fig.show()
```
